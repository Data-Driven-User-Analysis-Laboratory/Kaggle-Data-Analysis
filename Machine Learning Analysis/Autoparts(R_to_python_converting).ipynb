### 회귀분석

```python
import statsmodels.api as sm
y = car1.loc[:,"c_thickness"]
x = car1.loc[:,:"highpressure_time"]
lm = sm.OLS(y,x).fit()
```

```python
lm.summary()
```

- 모든 변수의 p-value가 0이므로 모든 변수가 회귀식에 통계적으로 유의하다고 판단할 수 있음.

```python
lm.tvalues
```

```python
import numpy as np
F_test = lm.f_test(np.identity(6))
```

```python
F_test
```

- 모델의 p-value가 0.05보다 작으므로 통계적으로 유의한 모델이라고 할 수 있음.

```python
y_predict1 = lm.predict(x)
```

### 다중공선성 계수 측정

```python
data = []

for i, a in enumerate(alpha_ridge):
  model_ridge = Ridge(alpha=alpha_ridge[i])
  model_ridge.fit(train_x,train_y)
  data.append(pd.Series(np.hstack([model_ridge.intercept_, model_ridge.coef_])))
  pred_y_ridge = model_ridge.predict(test_x)

df_ridge = pd.DataFrame(data, index = alpha_ridge).T
```

### Ridge model과 Lasso model의 alpha값 변화에 따른 결정계수 차이

- ridge의 경우 어떤 alpha값을 선택해도 변수 전체적인 계수의 값이 존재함
- lasso의 경우 alpha값이 낮으면 상수항을 제외한 모든변수의 계수가 0으로 출력

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(glm_train_x, glm_train_y)
```

```python
glm_y_pred = logreg.predict(glm_test_x)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(glm_test_x, glm_test_y)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(glm_test_y, glm_y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(glm_test_y, logreg.predict(glm_test_x))
fpr, tpr, thresholds = roc_curve(glm_test_y, logreg.predict_proba(glm_test_x)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
